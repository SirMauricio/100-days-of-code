###### PUNTO 3

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
y = df["Legendario"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=50)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", accuracy)
```

```python
df['Legendario'] = df ['Legendary'].apply(lambda x: 1 if x >= True else 0)
df['Legendario'].value_counts()
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(100, 50))
plot_tree(model, filled=True, feature_names=X.columns, class_names=['Legendario','No Legendario'] )
plt.title("Árbol de decisión")
plt.show()
```

###### PUNTO 4

```python
X_cluster = df[['Attack','Defense','Speed']]

from sklearn.cluster import KMeans
```

```python
modelo_kmeans = KMeans(n_clusters=3, random_state=50)
df['Grupo'] = modelo_kmeans.fit_predict(X_cluster)

df[['Grupo','Attack','Defense','Speed']].head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Attack'], df['Defense'],df['Speed'], c=df['Grupo'], cmap='rainbow')
plt.xlabel('Tanque')
plt.ylabel('Balanceado')
plt.title('Veloces')
plt.grid(True)
plt.colorbar(label='Grupo asignado')
plt.show()
```
